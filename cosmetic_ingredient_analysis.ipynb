```python
import pandas as pd

df = pd.read_csv('cosmetics.csv')

print(df.head())

print("\nNumber of Moisturizers in this dataset:")
print(df[df['Label'] == 'Moisturizer'].shape[0])
```

```python
luxury_brands = df[df['Price'] > 50].sort_values(by='Price', ascending=False)

print("Top 10 Luxury Brands in your Dataset:")
print(luxury_brands[['Brand', 'Name', 'Price']].head(10))

luxury_brands.to_csv('luxury_skincare_report.csv')
```
